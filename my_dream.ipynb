## How to use it

```python
# ADD THE LIBRARY
import pandas as pd
from vizzu import Chart, Data, Slide
```

```python
# CREATE THE CHART
# There is no need to specify the width and height right now. We could specify it later, when needed!
vc = Chart()
```

```python
## LOAD AND EXPLORE SOME DATA
# Read the excel file
df = pd.read_excel("data/CheeseSales.xlsx", sheet_name="sales")
# See what's inside
df
```

```python
## ADD INITIAL DATA TO THE CHART
vc_data = Data()
vc_data.add_df(df)
# Create the chart and add the data
vc.set_data(vchart_data) # A separate method, to be consistent with add_record(s)
```

```python
# ADD CONTENT
# Slide #1
slide = Slide(title="Cheese sales per hour", x="Date", y="Price") #default values ...
vc.add_slide(slide)
```

```python
## SEE HOW IT'S COMING ALONG
vc.show()
```

```python
# ADD MORE CONTENT
# Add a new record
# Should alert if not all the columns are met
vc.add_record({"Day":"2023-09-04", "Time":"08:00", "Type":"Mantecoso", "Weight":0.5, "Price":4.0}) 
# Slide #2
slide = Slide(title="Cheese sales per hour", x="Hour", y="Price", on_slide=10, on_animation=2)
vc.add_slide(slide)
# Slide #3
slide = Slide(title="Cheese weight sales per type", graph="pie", angle="Type", by="Weigth")
vc.add_slide(slide)
```

```python
## SEE HOW IT'S COMING ALONG
# The chart size is defined on rendering, not when it's defined as an object
vc.show(width="100%")
```

```python
## LOAD AND EXPLORE SOME ADDITIONAL DATA
# Read the excel file
df2 = pd.read_excel("data/CheeseSales.xlsx", sheet_name="price")
# See what's inside
df2
```

```python
# ADD THE NEW DATA FOR THE FUTURE CHARTS
vc_data2 = Data()
vc_data2.add_df(df2)
vc.set_data(vchart_data2)
```

```python
# Create the chart and add the data
slide = Slide(title="Cheese types and price", graph="pie", angle="Type", by="Price/Weight")
vc.add_slide(slide)
```

```python
## SEE HOW IT'S COMING ALONG
vc.show(width="800px", height="600px")
```
